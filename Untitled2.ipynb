```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/reklama_sales.csv')
df.head(5)
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=df)
plt.show()
```

```python
df = df.drop('Radio', axis=1)
df = df.drop('Newspaper', axis=1)
```

```python
from sklearn.model_selection import train_test_split
train_set , test_set= train_test_split(df, test_size=0.2, random_state=42)
X_train = train_set.drop('Sales', axis=1)
y = train_set['Sales'].copy()

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

num_pipeline = Pipeline([
          ('imputer', SimpleImputer(strategy='median')),
          ('std_scaler', StandardScaler())             
])
```

```python
X_prepared = num_pipeline.fit_transform(X_train)
```

```python
X_prepared[0:5,:]
```

```python
from sklearn.linear_model import LinearRegression

LR_model = LinearRegression()
```

```python
LR_model.fit(X_prepared, y)
```

```python
# tasodifiy 5 ta qatorni ajratib olamiz
test_data = X_train.sample(5)
test_data
```

```python
# yuqoridagi qatorlarga mos keluvchi narxlarni ajratib olamiz (biz aynan shu qiymatlarni bashorat qilishimiz kerak)
test_label = y.loc[test_data.index]
test_label
```

```python
test_data_prepared = num_pipeline.transform(test_data)
test_data_prepared
```

```python
predicted_data = LR_model.predict(test_data_prepared)
predicted_data
```

```python
pd.DataFrame({'Prognoz':predicted_data, 'Real baxosi': test_label})
```

```python
X_test = test_set.drop('Sales', axis=1)
X_test
```

```python
y_test = test_set['Sales'].copy()
```

```python
X_test_prepared = num_pipeline.transform(X_test)
```

```python
y_predicted = LR_model.predict(X_test_prepared)
```

```python
from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y_test, y_predicted)
# RMSE hisoblaymiz
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF_model = RandomForestRegressor()
RF_model.fit(X_prepared, y)
```

```python
y_predicted = RF_model.predict(X_test_prepared)
lin_mse = mean_squared_error(y_test, y_predicted)
# RMSE hisoblaymiz
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```
